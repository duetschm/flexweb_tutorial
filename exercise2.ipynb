## 2.1) Plot the plume on a map

First let's plot the whole plume on a map to see where it goes. Select the timestep you want to plot (e.g. 120 corresponds to 5 days if you have hourly output).

```python
# time step
timestep = 120

# open flexpart
ds, heights_exp = open_flexpart("FLEXPART_grid.nc")

# open topography file
etopo = open_topography("etopo5.nc")
```

```python
# colormap and levels
cmap = colormap_vintage()

logmax = int(np.floor(np.log10(np.max(ds["spec001_mr"]))))

levels = 10 ** np.arange(logmax - 6, logmax - 2 + 1e-10, 0.2)
norm = mcolors.LogNorm(vmin=10**(logmax-6), vmax=10**(logmax-2))
```

```python
# plot map
fig = plot_map(ds.isel(time=[timestep]), cmap, levels, norm)
```

**Notes on the result**: The unit "ng/m3" refers to the mass of released aerosols per m$^3$ above each grid cell. Here it is averaged over the output heights.
